```python
#read CSV and turn into a dataframe
csvfile = "global_weather_data_clean.csv"
df = pd.read_csv(csvfile)
df.head()
```

```python
#clean DV for html file
pd.set_option('display.max_colwidth', -1)
open('data.html', 'w').write(df2.to_html(na_rep = "", index = False))
```
